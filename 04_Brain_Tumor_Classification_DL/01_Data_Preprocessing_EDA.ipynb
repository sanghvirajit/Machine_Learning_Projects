### Visualizing training dataset

```python
### Function to plot images
def plot_images(path, title):
    
    from tensorflow.keras.preprocessing.image import load_img
    n = len(os.listdir(path))
        
    j = 10
    i = 10
    plt.figure(figsize=(20, 20))  
    
    for index, filename in enumerate(os.listdir(path)):
        
        fullname = f'{path}/{filename}'
        img = load_img(fullname)
        x = np.array(img)
        
        plt.subplot(i, j, index+1)
        
        plt.imshow(x)

        plt.xticks([])
        plt.yticks([])
                
        plt.suptitle('Tumor: {}'.format(title))
    plt.show()
```

```python
path = './Data/training/yes'

plot_images(path, "yes")
```

```python
path = './Data/training/no'

plot_images(path, "No")
```

```python
### function to show the balance-imbalance dataset
def plot(path):
    
    path_yes = f'{path}/yes'
    path_no = f'{path}/no'
    yes = len(os.listdir(path_yes))
    no = len(os.listdir(path_no))
    
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    langs = ['Yes', 'No']
    students = [yes,no]
    ax.bar(langs,students)
    plt.show()
```

```python
plot("./Data/training")
```

```python
plot("./Data/validation")
```

```python
plot("./Data/test")
```
