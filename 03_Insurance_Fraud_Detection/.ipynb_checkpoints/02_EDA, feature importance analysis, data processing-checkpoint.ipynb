# Mutual Information score between categorical columns

**We can figure out the relationship between categorical columns and our target variable by using mutual information.**

```python
from sklearn.metrics import mutual_info_score

def mutual_info_salary_score(series):
    return mutual_info_score(series, df.fraud_reported)
```

```python
mi = df[categorical].apply(mutual_info_salary_score)

mi.sort_values(ascending=False)
```

**incident_severity has the strongest relationship with the target variable, as compared to the rest of the categorical variables and we figured that out too during EDA**
